```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
import matplotlib.pyplot as plt
```

```
def hist(image):
  
  cv2_imshow(image)
  
  #cv2_imshow(image[:,:,0])
  hist_red = cv2.calcHist([image[:,:,0]],[0],None,[256],[0,256])
  plt.figure()
  plt.title("Red")
  plt.xlabel("Bins")
  plt.ylabel("# of Pixels")
  plt.plot(hist_red)
  plt.xlim([0, 256])

  
  
  #cv2_imshow(image[:,:,1])
  hist_green = cv2.calcHist([image[:,:,1]],[0],None,[256],[0,256])
  plt.figure()
  plt.title("Green")
  plt.xlabel("Bins")
  plt.ylabel("# of Pixels")
  plt.plot(hist_green)
  plt.xlim([0, 256])


  #cv2_imshow(image[:,:,2])
  hist_blue = cv2.calcHist([image[:,:,2]],[0],None,[256],[0,256])
  plt.figure()
  plt.title("Blue")
  plt.xlabel("Bins")
  plt.ylabel("# of Pixels")
  plt.plot(hist_blue)
  plt.xlim([0, 256])
  
  
  
  gray = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2GRAY)
  #cv2_imshow(gray)
  hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
  plt.figure()
  plt.title("Grayscale Histogram")
  plt.xlabel("Bins")
  plt.ylabel("# of Pixels")
  plt.plot(hist)
  plt.xlim([0, 256])

  
  return (hist_red, hist_green, hist_blue, hist)
```

```
def compare_histograms(image1, image2):
  red, green, blue, gray=hist(cv2.imread('/content/tiger.jpg'))
  red2, green2, blue2, gray2=hist(cv2.imread('/content/tiger (2).jpg'))
  
  gray_=cv2.compareHist(gray, gray2, cv2.HISTCMP_CORREL)
  print(' GrayScale: ', gray_)
  
  
  red_comp=cv2.compareHist(red, red2, cv2.HISTCMP_CORREL)
  green_comp=cv2.compareHist(green, green2, cv2.HISTCMP_CORREL)
  blue_comp=cv2.compareHist(blue, blue2, cv2.HISTCMP_CORREL)

  print(' Red: ', red_comp, '\n Green: ', green_comp, '\n Blue: ', blue_comp)

  
  
```

```
compare_histograms(cv2.imread('/content/tiger.jpg'), cv2.imread('/content/tiger (2).jpg'))
```
