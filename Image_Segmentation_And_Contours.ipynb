## Contours

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
image= cv2.imread('/content/spoon and plate.jpg')
image=cv2.resize(image, (500, 500))
cv2_imshow(image)
```

```
height, width= image.shape[:2]
```

```
end_row= int(height*0.8)
```

```
image=image[: end_row, :]
cv2_imshow(image)
```

```
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```

edged=cv2.Canny(gray, 300, 50)
cv2_imshow(edged)
```

```
_, contours, hierarchy=cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cv2_imshow(edged)
```

```
print('Number of contours' , str(len(contours)))
```

```
cont_image=cv2.drawContours(image, contours, -1, (0,0,0), 3)
cv2_imshow(cont_image)
```

```
for contour in contours:
   
    (x,y,w,h) = cv2.boundingRect(contour)
    cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 2)
    
cv2_imshow(image)
```

```
for contour in contours:
   
    (x,y,w,h) = cv2.boundingRect(contour)
    print(x,y, w,h)
```

```
image= cv2.imread('/content/spoon and plate.jpg')
image=cv2.resize(image, (1000, 1000))
height, width= image.shape[:2]

start_row=int(height*0.1)

end_row=int(height*0.8)
end_col=int(width*0.175)

image=image[start_row: end_row, :end_col]
cv2_imshow(image)
```

```
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
```

```
gaussian=cv2.GaussianBlur(gray, (5, 5), 0)
edged=cv2.Canny(gaussian, 250, 122)
cv2_imshow(edged)
```

```
_, contours, hierarchy=cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2_imshow(edged)
```

```
print('Number of contours' , str(len(contours)))
```

```
cv2.drawContours(image, contours, -1, (0,0,0), 3)
cv2_imshow(image)
```

```
for contour in contours[:9]:
    (x,y,w,h) = cv2.boundingRect(contour)
    cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 2)
```
