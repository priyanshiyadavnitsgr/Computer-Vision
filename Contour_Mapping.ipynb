## Image Preprocessing

```
def image_preprocessing(image_path):
  ref_image=Image.open(image_path, 'r')
  ref_image=ref_image.resize((300,300))
  #ref_image = cv2.cvtColor(ref_image, cv2.COLOR_BGR2GRAY)
  img_unsharp=ref_image.filter(ImageFilter.UnsharpMask(150)).filter(ImageFilter.CONTOUR)
  return img_unsharp
  
```
